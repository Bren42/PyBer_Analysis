```python
import matplotlib.pyplot as plt
```

```python
savings=[0,100,150,200,300,500,300,350,500,600,750,1000]
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(months,savings) #x values in the first order, y in the second
plt.xlabel('Month')
plt.ylabel('Savings($)')
plt.title('Saving account results')
plt.show()
```
