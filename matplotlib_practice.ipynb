```python
#import the dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in us dollars accumulated for each month.

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis)
```

```python
#create a plot with ax.plt (this is an object oriented method versus MATLAB in previous cells)

fig,ax= plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot and add a label to the legend
plt.plot(x_axis,y_axis,marker="D",color="green",linewidth=2,label='Boston')

#create a label for each axis
plt.xlabel("Date")

plt.ylabel('Fare($)')

#set the y limit between 0 and 45.

plt.ylim(0,50)


#create a title

plt.title('PyBer Fare by Month')



#add the legend

plt.legend()
```

```python
#set the x-axis to a list of strings for each month
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in us dollars accumulated for each month.

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot

plt.bar(x_axis,y_axis,color='green',label='Boston')

#create labels for x and y axes

plt.xlabel("Date")
plt.ylabel("Fare($)")

#create a title

plt.title("PyBer Fare by Month")

#add the legend
plt.legend()
```

```python
#create the plot for a horizontal bar chart

plt.barh(x_axis,y_axis,color="magenta",label="Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()

plt.legend()
```

```python
#create the charts using the object oriented method, set the axis to a string
fig,ax=plt.subplots()

ax.barh(x_axis,y_axis,color="steelblue",label="Chicago")

ax.set_xlabel("date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```
