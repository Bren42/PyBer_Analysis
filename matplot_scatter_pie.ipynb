```python
%matplotlib inline

#import the dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#create a scatter plot with MATLAB method(lowercase 'o' in the chart below tells the system to change from a line chart to a scatter chart

plt.plot(y_axis, x_axis,'o',color='red',label='Chicago')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(5)
plt.legend()
```

```python
#create a bubble chart
#plt.scatter(x_axis, y_axis)
plt.scatter(x_axis, y_axis, s = [i * 4 for i in y_axis])
# this is the comprehension flow for the above code "new_list = [expression for item in list if conditional]"

#create a larger bubble to make the chart more easy to read, this is another way to adapt the bubble sizing
# y_axis_larger=[]
# for data in y_axis:
#     y_axis_larger.append(data*3)
```

```python
#object oriented scatter plot
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#object oriented bubble chart
fig, ax=plt.subplots()

ax.scatter(y_axis, x_axis,s=[i * 5 for i in y_axis],linewidth=2,edgecolor='black',alpha=0.5,color='skyblue',label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.invert_yaxis()
ax.legend()
ax.set_xlim(0)
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show
```

```python
plt.subplots(figsize=(8,8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#create a pie chart using the object oriented interface

fig, ax=plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show
```

```python
#pie chart uisng object oriented where I have increased the size and colors, shadows and moved jan to top by adjusting angle.
fig1,ax1 = plt.subplots(figsize=(8,8))

explode=[0,0,0.3,0,0,0,0.3,0,0,0,0,0]

colors=["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]



ax1.pie(y_axis,labels=x_axis,startangle=90,explode=explode,shadow=True,autopct='%1.1f%%',counterclock=False,colors=colors)
plt.show
```
